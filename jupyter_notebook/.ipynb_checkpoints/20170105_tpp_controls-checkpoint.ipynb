1. Adding ACGU seems to dramatically improve growth.  I know these graphs don't really show it, but the counts/sec for the positive and negative controls were much more similar with ACGU than without it.

2. With ACGU, I can grow the cells for the same amount of time I normally do, which is 9h.  That's good, because it means I won't have to change my screening schedule (which probably wouldv'e been difficult).

3. `tpp rxb 11,1` has even weaker activity than `amm rxb 11,1`, despite the fact that the thiamine aptamer binds it ligand much tighter than the ammeline aptamer does.  Still, the aptamer is functional, and that's a good sign.
