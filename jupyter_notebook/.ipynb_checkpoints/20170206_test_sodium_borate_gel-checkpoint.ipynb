# Test sodium borate gel

I found a reference claiming that sodium borate is a much better electrophoresis buffer than TBE and TAE, because it doesn't heat up as much and can therefore be subjected to higher voltages.  I want to see if this works for me.

## Prepare 20x sodium borate (SB)

- 8.0 g NaOH
- 48.75 g boric acid
- water to 1L

## Prepare 1x sodium borate (SB)

- 50 mL 20x sodium borate
- water to 1L

## Cast a gel

- 0.5 g agarose
- 50 mL 1x SB
- 5 μL GelRed

## Run a ladder

- Load 4 μL GeneRuler 1 kb DNA ladder
- Run at 50.0 W for 12 min.

Condition | Voltage (V) | Current (mA) | Power (W) | Temperature (°C)

before    | 361         | 139          | 50.0      | 21.6
after     | 322         | 155          | 50.0      | 46.2
