# Adenine/guanine controls

The purpose of this experiment is two-fold:

1. Determine how long I need to grow cells in MEKAG media (MOPS + EZ supplement + K2HPO4 + adenine + guanine) for the sensors to give the largest fold-change.

2. Determine whether I'd rather use adenine or guanine as a ligand.

To do this, I cloned the adenine and guanine aptamers into the four GFP and RFP spacers I've been using.  I then grew cells with these sensors in MEKAG media for 12h, taking timepoints every hour from 7h on.  I then analyzed the time point by flow cytometry.

## Results

1. The controls aren't as insensitive to ligand as I'd like.  Especially in the un-normalized data, there is a ~1.5 - 2.0x difference between the controls for all four spacers.  The fact that this difference is smaller in the normalized data suggests that it may be tied to a growth defect.

2. The cells didn't grow especially well in this media.  You can't see this is in the traces or the fold changes, but it wasn't until 11h that the cultures really started to look saturated (I know that's very qualitative).

3. There was also a lot of contamination between wells, which is always exacerbated when some wells have significantly fewer cells than others.  I expect that most of the bimodal or messy peaks in this data can be attributed to contamination.

4. What is going on with rfp add rxb 11,1?  If that's a real result, it's incredible, especially considering that none of the other sensors in this experiment work very well at all.  But I'm pretty suspicious that I accidentally mixed up some wells of something.  I should try to repeat this.

5. The add aptamer looks more promising than the gua aptamer, based on the data for gfp1.  I don't want to conclude anything from the rfp1 data, because those data are such outliers.  For the other two spacers, they both look inactive.  So this is a weak conclusion, but if I were going to do a screen, I'd do it against the add aptamer.

I'm not comfortable starting a screen based on these data, primarily because the controls are so sensitive to ligand.
